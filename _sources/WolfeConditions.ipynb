# Lema 3.1

Suponga que $f:\mathbb{R}^n\to\mathbb{R}$ es continua y diferenciable. Sea $p_k$ una direccion de descenso en $x_k$, y asuma que $f$ es acotado por debajo por los rayos $\ell_{\alpha} =\{x_k+\alpha p_k\mid \alpha > 0 \}$. 
Entonces para  $0< c_1< c_2 < 1$, entonces existen intervalos de tamaños de paso que satisface las condiciones de Wolf y las condiciones fuertes de Wolf.

## Demostracion

Consideremos

$$
  \phi(\alpha)=f(x_k+\alpha p_k)
$$

Como $f$ es acotada $\phi $ tambien lo es. Definimos 

$$
  l(\alpha )=f(x_k)+\alpha c_1\nabla f(x_k)^Tp_k
$$ 

Como $l(\alpha)$ no esta acotada por debajo necesariamente existe $\alpha^*=\min\{\alpha\mid f(x_k+\alpha p_k)=\phi (\alpha)\}$ tal que 

$$
  f(x_k+\alpha^*p_k)=f(x_k)+\alpha ^*c_1 \nabla f(x_k)^Tp_k\ \ (⋆)
$$ 

Luego, $ℓ_\alpha$ es un conjunto convexo $∀\alpha>0$, por lo tanto al tomar 
$\alpha'\in(0,\alpha^*)$ entonces

$$
  f(x_k+\alpha ^*p_k)\leq f(x_k)+\alpha ^*c_1 \nabla f(x_k)^Tp_k
$$

Teniendo asi la condición de suficiente decrecimiento. Luego, para el segundo criterio usaremos el Teorema del valor medio para funciones multivariables. 
Entonces, para algun $\alpha^{**}\in (0,\alpha^*)$

$$
  f(x_k+\alpha^*p_k)=f(x_k)+\alpha^* \nabla f(x_k+\alpha^{**}p)p^T (⋆⋆)
$$

Por $(\star)$ tenemos que 

$$
  f(x_k+\alpha^*p_k)-f(x_k)=\alpha^*c_1\nabla f(x_k)^Tp_k
$$

,De manera similar con $(\star\star)$, e igualando tenemos que 

$$
  \alpha^* \nabla f(x_k+\alpha^{**} p ) p^T=\alpha^*c_1 \nabla f(x_k)^Tp_k
$$

y para algun $1 > c_2 > c_1$ tenemos que 
$$ 
  \nabla f(x_k+\alpha^{**} p)p^T=c_1\nabla f(x_k)^Tp_k \geq c_2 \nabla f(x_k)^Tp_k
$$

Como ambos son negativos tenemos las condiciones de Wolf, ahora para obtener las fuertes debemos aplicar el valor absoluto a la segunda condicion, por tanto se invertira la desigualdad.Entonces teneemos que 

 $$
   \text{abs}(\nabla f(x_k+\alpha^{**}p)p^T)\leq c_2 \text{abs}(\nabla f(x_k)^Tp_k)
 $$

 Teniendo asi las condiciones fuertes de Wolf.
